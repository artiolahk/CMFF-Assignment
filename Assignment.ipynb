```python
# Step 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import yfinance as yf  # Yahoo Finance API library

# Step 2: Define the equity symbol and time period
equity_symbol = "BATS.L"  
start_date = "2021-11-20"
end_date = "2023-11-20"

# Step 3: Download historical daily data using Yahoo Finance API
equity_data = yf.download(equity_symbol, start=start_date, end=end_date)

# Step 4: Check the downloaded data
print(equity_data.head())

# Step 5: Plot the equity price movement
plt.figure(figsize=(10, 6))
plt.plot(equity_data['Adj Close'], label=f'{equity_symbol} Adj. Closing Price')
plt.title(f'{equity_symbol} Price Movement ( {start_date} to {end_date} )')
plt.xlabel('Date')
plt.ylabel('Equity Price (USD)')
plt.legend()
plt.show()
```

```python
daily_returns = equity_data ['Adj Close'].pct_change ()
daily_returns.plot()
```
